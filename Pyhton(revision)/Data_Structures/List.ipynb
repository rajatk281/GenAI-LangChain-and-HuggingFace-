# List 
It stores ordered mutable(it can be changed) collection of multiple values

```python
nums = [1, 2, 3, 4]
names = ["raj", "amit", "neha"]
mixed = [1, "hello", True, 3.5]
```

#### Common List Operations

```python
nums.append(40)     # add at end
nums.insert(1, 15)  # add at index
nums.remove(40)     # remove by value
nums.pop()          # remove last
nums.pop(0)         # remove by index
nums.sort()         # sort list
nums.reverse()      # reverse list
```

#### Accessing List Elements

```python
nums = [10, 20, 30, 40]

nums[0]     # 10
nums[-1]    # 40
nums[1:3]   # [20, 30]
```

#### Looping Through a List

```python
for num in nums:
    print(num)

# With Index

for i, val in enumerate(nums):
    print(i, val)
```

#### List Comprehension
A short and clean way to make a list 

```python
# Syntax
new_list = [item+1 for item in nums]

# Normal loop vs List Comprehension

# Normal way
squares = []
for i in range(5):
    squares.append(i * i)

# List comprehension
squares = [i * i for i in range(5)]

#With condition 
evens = [x for x in range(10) if x % 2 == 0]



```
