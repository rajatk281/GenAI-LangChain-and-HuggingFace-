#### Common List Operations

```python
nums.append(40)     # add at end
nums.insert(1, 15)  # add at index
nums.remove(40)     # remove by value
nums.pop()          # remove last
nums.pop(0)         # remove by index
nums.sort()         # sort list
nums.reverse()      # reverse list
```

#### Accessing List Elements

```python
nums = [10, 20, 30, 40]

nums[0]     # 10
nums[-1]    # 40
nums[1:3]   # [20, 30]
```

#### Looping Through a List

```python
for num in nums:
    print(num)

# With Index

for i, val in enumerate(nums):
    print(i, val)
```

#### List Comprehension
A short and clean way to make a list 

##### [x for x in iterable]
##### [x for x in iterable if condition]
##### [x*y for x in iterable]

##### Syntax
###### new_list = [expression for item in iterable]

```python
# Normal loop vs List Comprehension

# Normal way
squares = []
for i in range(5):
    squares.append(i * i)

# List comprehension
squares = [i * i for i in range(5)]

# With condition 
evens = [x for x in range(10) if x % 2 == 0]

# String Processing 
names = [" Raj ", "Amit", " neha "]
clean_names = [name.strip().lower() for name in names]
print(clean_names)

# Nested List Comprehension
matrix = [[1,2], [3,4], [5,6]]
flat = [num for row in matrix for num in row]
print(flat)
```
