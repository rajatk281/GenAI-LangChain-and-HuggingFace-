# Dictionary

A dictionary is a collection data type in Python that stores data in key : value pairs.

Properties:

Unordered (insertion order preserved from Python 3.7+)

Mutable

Keys are unique

Keys must be immutable

Values can be any data type

```python
student = {
    "name": "Rajat",
    "age": 21,
    "course": "CSE"
}
```

#### Creating a Dictionary

```python
# Empty Dictionary
d = {}

# Using dict() constructor
d = dict(a=1, b=2)

# Mixed Data Types
d = {1: "one", (2,3): "tuple", "list": [1,2,3]}
```

#### Accessing Values

```python
d = {"a": 10, "b": 20}

print(d["a"])       # 10
print(d.get("b"))   # 20
print(d.get("c"))   # None (safe)
```

#### Adding & Updating Elements

```python
d = {"a": 1}

d["b"] = 2       # add
d["a"] = 10      # update
```

#### Removing Elements

```python
d = {"a": 1, "b": 2, "c": 3}

d.pop("b")        # removes key b
d.popitem()       # removes last inserted item
del d["a"]        # deletes key
d.clear()         # removes all items
```

```python
d = {"a": 1, "b": 2}

d.keys()       # dict_keys(['a','b'])
d.values()     # dict_values([1,2])
d.items()      # dict_items([('a',1),('b',2)])
```

#### Looping Through Dictionary

```python
# Loop through keys
for key in d:
    print(key, d[key])

# Loop through values
for val in d.values():
    print(val)

# Loop through key-value pairs
for k, v in d.items():
    print(k, v)
```

#### Dictionary Comprehension

```python
squares = {x: x*x for x in range(1, 6)}

# With condition 
even = {x: x for x in range(10) if x % 2 == 0}
```

#### Nested dictionaries

```python
students = {
    "s1": {"name": "A", "age": 20},
    "s2": {"name": "B", "age": 21}
}

print(students["s1"]["name"])
```

Allowed as keys:  int, float, string, tuple (immutable)

Not allowed: list, set, dict

```python
d = {(1,2): "ok"}   # Correct
d = {[1,2]: "no"}  # TypeError
```

```python
# Frequency count example 
s = "banana"
freq = {}

for ch in s:
    freq[ch] = freq.get(ch, 0) + 1
    
print(freq)
```

#### Mutable values allowed

```python
d = {"a": [1,2]}
d["a"].append(3)   # allowed
```
