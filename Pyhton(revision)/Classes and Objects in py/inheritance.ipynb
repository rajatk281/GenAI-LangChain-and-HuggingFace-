# Inheritance

```python

## Base class 1
class Shirt:
    def __init__(self, shirtColor):
        self.shirtColor = shirtColor

## Base class 2
class Pant:
    def __init__(self, pantColor):
        self.pantColor = pantColor
        
## Derived Class

class Student(Shirt, Pant):
    def __init__(self, name, shirtColor, pantColor ):
        self.name = name
        Shirt.__init__(self, shirtColor)
        Pant.__init__(self, pantColor)
        
    def speak(self):
        return f'{self.name} wears {self.shirtColor} shirt and {self.pantColor} pant'
    
student1 = Student('Rahul', 'blue', 'black')
student1.speak()
```
