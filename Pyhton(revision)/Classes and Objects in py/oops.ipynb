```python
## modelling school bag 

## Defining a class for school bag 

class Bag:
    def __init__(self, student, books, capacity):
        self.student = student 
        self.books = books
        self.capacity = capacity
        
    def putBooks(self, newBooks):
        self.newBooks = newBooks
        self.books+=newBooks
        if self.books<=self.capacity:
            print(f'{newBooks} books are kept in the bag now total books in the bag are {self.books}')
        else:
            print(f"{self.books} books are reaching the capacity limit")
            self.books-=newBooks
            
    def removeBooks(self, removedBooks):
        self.removedbooks = removedBooks
        self.temp = self.books
        self.books -= removedBooks
        if self.books<0:
            print(f'Bag contains only {self.temp} books')
        else:
            print(f'{removedBooks} books has been removed from the bag now {self.books} books are there in the bag')
    def getBooks(self):
        return self.books
    

bag1 = Bag('Rahul', 5, 10)
print(bag1.student)
print(bag1.books)
print(bag1.capacity)

bag1.putBooks(3)
bag1.getBooks()
bag1.removeBooks(5)
bag1.putBooks(8)
bag1.getBooks()
bag1.removeBooks(4)
    
```
