```python
# Reading a file line by line  
with open("Example.txt", "r") as file:
    for line in file:
        print(line.strip())
```

```python
# Writing a line in file (Over writing)
with open('Example.txt', 'w') as file:
    file.write("Content written successfully \n")
    file.write("Next line \n")
```

```python
# Writing a line in file (Without Over writing)
with open('Example.txt', 'a') as file:
    file.write("COntent without overwrite \n")
    file.write("Next line 2\n")
```

```python
## Writing a list of line 
lines = ['line 1 \n', 'line 2 \n', 'line 3 \n']
with open("Example.txt", 'a') as file:
    file.writelines(lines)
```
