```python
# Reading a file 
with open("Example.txt", "r") as file:
    for line in file:
        print(line.strip())
```

```python
# Writing a file (Over writing)
with open('Example.txt', 'w') as file:
    file.write("Content written successfully \n")
    file.write("Next line")
```

```python
# Writing a file (Without Over writing)
with open('Example.txt', 'a') as file:
    file.write("COntent without overwrite \n")
    file.write("Next line 2")
```
