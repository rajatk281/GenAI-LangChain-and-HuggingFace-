# filter()

filter() is a built-in function used to select elements from an iterable that satisfy a condition.

It filters out unwanted elements.

#### Syntax : ilter(function, iterable)

###### function -> returns True or False

###### iterable -> list, tuple, string, etc.

###### Returns a filter object (iterator)


```python
nums = [1, 2, 3, 4, 5]

even = filter(lambda x: x % 2 == 0, nums)
print(list(even))   # [2, 4]
```

#### filter() working 

Takes one element from iterable

Applies condition

Keeps element if result is True

Discards element if result is False

#### filter() with named function 

```python
def is_positive(x):
    return x > 0

nums = [-2, -1, 0, 1, 2]
print(list(filter(is_positive, nums)))
```

#### filter() returns an iterator 

```python
f = filter(lambda x: x > 2, [1, 2, 3, 4])

print(list(f))   # [3, 4]
print(list(f))   # []  (exhausted)
```

#### If function is null it removes falsy values

Falsy values:

0

False

None

""

[]

```python
data = [0, 1, "", "hi", None, 5]
print(list(filter(None, data)))
# [1, 'hi', 5]
```

#### filter() with strings 

```python
names = ["ram", "", "shyam", ""]

non_empty = list(filter(lambda x: x != "", names))
```

#### filter() vs list comprehension 

```python
# Using filter 
evens = list(filter(lambda x: x % 2 == 0, nums))

# Using list comprehension 
evens = [x for x in nums if x % 2 == 0]
```
