# Map()

map() is a built-in function used to apply a function to every element of an iterable (list, tuple, etc.) and return a map object (iterator).

map(function, iterable)


#### Basic examples

```python
nums = [1, 2, 3, 4]

result = map(lambda x: x * 2, nums)
print(list(result))   # [2, 4, 6, 8]
```

#### map() working 

Takes one element from iterable

Applies the function

Stores the result

Moves to next element

#### map() with named function

```python
def square(x):
    return x * x

nums = [1, 2, 3]
print(list(map(square, nums)))
```

#### map() with Multiple Iterables

```python
a = [1, 2, 3]
b = [4, 5, 6]

result = map(lambda x, y: x + y, a, b)
print(list(result))   # [5, 7, 9]
```

#### map() vs loop 

```python
# Using Loop 
res = []
for x in nums:
    res.append(x*x)

# using map()
res = list(map(lambda x: x*x, nums))
```

#### map() returns an iterator

```python
m = map(lambda x: x+1, [1,2,3])

print(list(m))   # [2,3,4]
print(list(m))   # [] (already consumed)
```

#### map() with type conversion 

```python
nums = ["1", "2", "3"]
ints = list(map(int, nums))
```

#### map() with strings

```python
nums = ["1", "2", "3"]
ints = list(map(int, nums))
```

#### map() + Lambda 

```python
celsius = [0, 20, 30]
fahrenheit = list(map(lambda c: (c*9/5)+32, celsius))
```

#### map() with input()

```python
nums = list(map(int, input().split()))  # split() : split the elements by spaces 
```
